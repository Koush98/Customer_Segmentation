```python
# 6) Print the unique values in "KM" column.
df["KM"].unique()
```

```python
# 7) Replace "three","four" as 3,4,5 in the "Doors" column.
df["Doors"]=df["Doors"].replace({"three":3, "four":4, "five":5})
df["Doors"]
```

```python
# 8) How many missing values are there
df.isnull().sum()
```

```python
# 9) Fill the missing values by their mean, median, mode whatever is applicable.
df["Age"].fillna(df["Age"].mean(), inplace=True)
```

```python
df["KM"].fillna(df["KM"].mean(),inplace=True)
```

```python
df["FuelType"].fillna(df["FuelType"].mode()[0],inplace=True)
```

```python
df["HP"].fillna(df["HP"].mean(),inplace=True)
```

```python
df["MetColor"].fillna(df["MetColor"].mode()[0],inplace=True)
```

```python
# 10) Create a scatter plot betwwen "Age" and "Price".
plt.scatter(df["Age"],df["Price"],marker="*")
plt.title("Scatter Plot of Age & Price")
plt.xlabel("Age")
plt.ylabel("Price")
plt.show()
```

```python
# 11) Plot a histogram of "KM" with appropriate title and labels which will show thw values of "KM" in 
#  X axis with corresponding frequencies in Y axis. 
plt.hist(x=df["KM"])
plt.xlabel("Kilometer")
plt.ylabel("Frequency")
plt.show()
```

```python
# 12) Construct a density plot of "Age"
sns.displot(df["Age"])
plt.show()
```

```python
# 13) Create a box plot of "price"
sns.boxplot(df["Price"],color="red")
plt.xlabel("Price")
plt.ylabel("frequency")
plt.show()
```

```python
# 14) Create a count plot between "Fuel Type" and "Automatic" and 
#  study about what kind of Fueltype cars are having manual Gear Box

# Create count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='FuelType',hue='Automatic', data=df)
plt.title('Distribution of Manual Gearbox Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
